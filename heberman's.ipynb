```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values 
```

```python
x.shape
```

```python
y =  y.reshape(-1,1)
```

```python
y
```

```python
x.shape
```

```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(x,y , test_size = 0.21 , random_state = 34)
```

```python
x_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 16, p=2,weights = "distance", leaf_size =1)
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
# from sklearn.metrics import r2_score
# r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test , y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'n_neighbors' : list(np.arange(50)),'weights': ['distance'],'p':[1,2,3],'leaf_size': list(np.arange(50)) }
grid_search = GridSearchCV(estimator=clf, param_grid=parameters, scoring='accuracy', n_jobs=-1)
```

```python
grid_search.fit(x_train, y_train)
```

```python
best_params = grid_search.best_params_
```

```python
best_params
```

```python
clf2 = KNeighborsClassifier(leaf_size =1,n_neighbors=29, weights='distance',p=1)
clf2.fit(x_train , y_train)
```

```python
pred = clf2.predict(x_test)
```

```python
accuracy = accuracy_score(pred , y_test)
```

```python
for i in range (1,51):
    x_train,x_test , y_train ,y_test = train_test_split(x,y,test_size = 0.21, random_state=i)
        
    classifier = KNeighborsClassifier(n_neighbors = 16, p=2,weights = "distance", leaf_size =1)
    
    classifier.fit(x_train,y_train)
    test =classifier.predict(x_test)
    train  = classifier.predict(x_train)
    
    acc_test = accuracy_score(test , y_test)
    print(f"for n_neighbors {i}accuracy on test data is : ",acc_test)
        
    acc_train  = accuracy_score(train , y_train )
    print(f"for n_neighbors {i}accuracy on train data is : ",acc_train)
    print("******************************")
```
